```python
# Task 1: Trend of Web Series Released Per Year Goal: Use a Line Chart to visualize the trend of web series released over time.

# Trend of Web Series Released Per Year - Visualization

#I'll create a line chart to visualize the trend of web series releases per year based on the provided dataset.

## Analysis

#First, let's count the number of web series released each year:

#- 2017: 3 series
#- 2018: 4 series
#- 2019: 8 series
#- 2020: 3 series
#- 2021: 9 series
#- 2022: 12 series
#- 2023: 6 series

## Line Chart Description

#The line chart will show:
#- X-axis: Years (2017 to 2023)
#- Y-axis: Number of web series released
#- Data points connected by a line to show the trend

## Interpretation

#The trend shows:
#1. A gradual increase from 2017 (3) to 2019 (8)
#2. A slight dip in 2020 (3) possibly due to pandemic effects
#3. A sharp increase in 2021 (9) and 2022 (12) as streaming platforms expanded
#4. A decrease in 2023 (6) - though this might be incomplete data if collected mid-year

## Python Code to Generate the Chart

import matplotlib.pyplot as plt

# Data
years = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
series_count = [3, 4, 8, 3, 9, 12, 6]

# Create line chart
plt.figure(figsize=(10, 6))
plt.plot(years, series_count, marker='o', linestyle='-', color='b', linewidth=2)

# Add titles and labels
plt.title('Trend of Web Series Released Per Year', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Web Series Released', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Display the chart
plt.show()
```

```python
#5  Task 5: Relationship Between Number of Episodes and IMDB Rating Goal: Use a Scatter Plot to observe if web series with more episodes tend to have higher ratings.

# **Relationship Between Number of Episodes and IMDB Rating (Scatter Plot Analysis)**

## **Objective**
#To analyze whether **longer web series (more episodes)** tend to receive **higher IMDB ratings** using a **scatter plot** with trendline.


## **Key Insights from the Scatter Plot**

#✅ **No Strong Correlation Found**
#- The scatter plot shows **no clear trend** between episode count and rating.
#- Some **short series (6-8 episodes)** have **very high ratings (9.0+)**.
#- Some **longer series (12-14 episodes)** have **average ratings (6.5-7.5)**.

#🔍 **Notable Observations**
#- **Highest-rated series (9.3)**:
 # - *Criminal Justice* (6 episodes)
 # - *Mumbai Diaries* (14 episodes)
 # - *Leila* (11 episodes)
#- **Lowest-rated series (6.5)**:
 # - *Mirzapur* (12 episodes)
 # - *Breathe* (8 episodes)

#📉 **Conclusion**:
#- **Episode count does not guarantee higher ratings.**
#- **Quality matters more than quantity**—some short series outperform long ones.


## **Python Code for Scatter Plot**

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress

# Sample data (Episodes vs. IMDB Rating)
data = {
    "Episodes": [8, 12, 13, 13, 14, 14, 8, 13, 10, 13, 11, 6, 12, 10, 7, 14, 13, 14, 13, 10, 10, 13, 10, 10, 11, 14, 7, 6, 11, 11, 10, 11, 9, 6, 9, 13, 8, 11, 14, 8, 6, 8, 8, 6, 14, 10],
    "Rating": [8.8, 6.5, 6.8, 7.9, 7.1, 6.9, 6.5, 6.7, 8.6, 8.5, 7.1, 9.3, 7.0, 8.4, 6.9, 8.3, 8.2, 7.8, 6.6, 7.1, 6.9, 6.7, 7.9, 9.1, 9.1, 9.3, 6.5, 6.9, 9.1, 6.9, 7.1, 9.3, 8.3, 7.3, 6.7, 7.6, 8.0, 7.4, 6.7, 8.7, 8.0, 6.6, 8.6, 6.6, 6.9, 6.9]
}

df = pd.DataFrame(data)

# Calculate regression line
slope, intercept, r_value, p_value, std_err = linregress(df["Episodes"], df["Rating"])
regression_line = slope * df["Episodes"] + intercept

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df["Episodes"], df["Rating"], color='#1f77b4', alpha=0.7, s=100, edgecolor='black')

# Add regression line
plt.plot(df["Episodes"], regression_line, color='red', linestyle='--', label=f'Trendline (R² = {r_value**2:.2f})')

# Highlight top-rated series
top_rated = df[df["Rating"] >= 9.0]
plt.scatter(top_rated["Episodes"], top_rated["Rating"], color='gold', s=150, edgecolor='black', label='Top Rated (≥9.0)')

# Customize plot
plt.title("IMDB Rating vs. Number of Episodes", fontsize=14, pad=15)
plt.xlabel("Number of Episodes", fontsize=12)
plt.ylabel("IMDB Rating", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Annotate top-rated points
for _, row in top_rated.iterrows():
    plt.text(row["Episodes"] + 0.3, row["Rating"] - 0.1, f'{row["Rating"]}', fontsize=9, ha='left')

plt.tight_layout()
plt.show()
```

```python
#7 Task 7: Episodes Per Season Analysis Goal: Compute and compare the average number of episodes per season for each web series. Visualization: Bar Chart or Bubble Chart (Series on X-axis, Episodes/Season ratio as value).

# **Episodes Per Season Analysis**

## **Objective**
#To **compare the episode density** of web series by calculating the **average number of episodes per season** and visualizing it with a **bar chart** (or **bubble chart** for added dimensionality).


## **Key Insights**

### **1. High Episode Density (More Episodes/Season)**
#- **"Delhi Crime"**: **4.67 episodes/season** (14 episodes ÷ 3 seasons)
#- **"Aarya"**: **4.67 episodes/season** (14 episodes ÷ 3 seasons)
#- **"Special OPS"**: **4.33 episodes/season** (13 episodes ÷ 3 seasons)

### **2. Low Episode Density (Fewer Episodes/Season)**
#- **"Criminal Justice"**: **6 episodes/season** (6 episodes ÷ 1 season)
#- **"Four More Shots Please"**: **6 episodes/season** (6 episodes ÷ 1 season)
#- **"Leila"**: **3.67 episodes/season** (11 episodes ÷ 3 seasons)

### **3. General Trend**
#- **Crime/Thriller** series tend to have **higher episode density** (more episodes per season).
#- **Limited series (1 season)** often have **fewer episodes** but sometimes higher ratings.


## **Python Code for Bar Chart**

### **Step 1: Calculate Episodes/Season Ratio**
import pandas as pd

# Sample data (replace with full dataset)
data = {
    "Series": ["Delhi Crime", "Aarya", "Special OPS", "Criminal Justice", "Four More Shots Please", "Leila"],
    "Episodes": [14, 14, 13, 6, 6, 11],
    "Seasons": [3, 3, 3, 1, 1, 3],
    "IMDB Rating": [6.9, 8.3, 6.7, 9.3, 7.3, 9.3]
}

df = pd.DataFrame(data)
df["Episodes/Season"] = df["Episodes"] / df["Seasons"]

### **Step 2: Visualize with Bar Chart**
import matplotlib.pyplot as plt

# Sort by Episodes/Season
df = df.sort_values("Episodes/Season", ascending=False)

# Create bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(
    df["Series"],
    df["Episodes/Season"],
    color=plt.cm.YlGnBu(df["IMDB Rating"] / 10),  # Color by IMDB rating
    edgecolor="black"
)

# Add labels and title
plt.title("Average Episodes Per Season by Web Series", fontsize=14, pad=15)
plt.xlabel("Web Series", fontsize=12)
plt.ylabel("Episodes Per Season", fontsize=12)
plt.xticks(rotation=45, ha="right")

# Add data labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f"{height:.1f}", ha="center", va="bottom")


### **Alternative: Bubble Chart (Episodes/Season vs. Rating)**
plt.figure(figsize=(12, 6))
plt.scatter(
    df["Series"],
    df["Episodes/Season"],
    s=df["IMDB Rating"] * 50,  # Bubble size = IMDB Rating
    c=df["IMDB Rating"],
    cmap="YlGnBu",
    alpha=0.7,
    edgecolor="black"
)

plt.title("Episodes Per Season vs. IMDB Rating (Bubble Chart)", fontsize=14)
plt.xlabel("Web Series", fontsize=12)
plt.ylabel("Episodes Per Season", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.colorbar(label="IMDB Rating")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
